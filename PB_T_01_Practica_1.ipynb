<h2 style="color:#2E4053; font-family:Georgia; border-left:6px solid #2E86C1; padding-left:10px;">
üìä Caso 1: Costo de distribuci√≥n log√≠stica
</h2>


```python
# 1) Leer los valores desde la entrada del programa (una linea cada uno)
D=input('Ingrese el n√∫mero de destinos: ')
R=input('Ingrese la tarifa base de destino en USD: ')
F=input('Ingrese el factor de riesgo log√≠stico: ')

# 2) Convertir cada valor de texto obtenido en un numero decimal
D=float(D)
R=float(R)
F=float(F)

# 3) Calcular el costo total con la formula
C=D*R*(1+(0.05*F))

# 4) Imprimir el resultado con exactamente dos decimales
print(f'El costo total de la distribuci√≥n es ${C:.2f}')
```
