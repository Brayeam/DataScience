<h2 style="color:#2E4053; font-family:Georgia; border-left:6px solid #2E86C1; padding-left:10px;">
Caso 1: üìä Problema: Registro de inscripciones a un curso
</h2>


````python
# Funci√≥n para validar las inscricpiones
def validar_inscripcion(inscripcion):
    nombre, edad, sede = inscripcion

    """
    Generar la validaci√≥n de tipos
    eval√∫a:     - nombre  (cadena de texto)
                - sede    (cadena de texto)
                - edad    (n√∫mero entero)

    Valida rango de fecha permitido (16 - 120)

    Eval√∫a si el registro es valido

    """

    # validar tipos
    if not isinstance(nombre, str):
        return False
    if not isinstance(sede, str):
        return False
    if not isinstance(edad, int):
        return False

    # Validar rango de edad
    if not (16 <= edad <= 120):
        return False

    return True


# Funci√≥n para procesar inscripciones
def procesar_inscripciones(lista):
    inscripciones_unicas = set()
    registro_por_sede = {}

    """
    Utiliza previa funci√≥n validar_inscripcion para verificar inscripcion previo a procesamiento
    generar un conjunto que lleve las inscripciones √∫nicas (combinacion nombre + sede √∫nica sin repetici√≥n)
    ignorar inscripciones inv√°lidas
    ignorar inscripciones duplicadas
    Almacenar inscripciones validas en diccionario
    Generar resumen final

    """

    for inscripcion in lista:
        if not validar_inscripcion(inscripcion):
            print("Inscripci√≥n inv√°lida:", inscripcion)
            continue

        nombre, edad, sede = inscripcion
        clave_unica = (nombre, sede)

        if clave_unica in inscripciones_unicas:
            print("Inscripci√≥n duplicada ignorada:", inscripcion)
            continue

        inscripciones_unicas.add(clave_unica)

        if sede not in registro_por_sede:
            registro_por_sede[sede] = []
        registro_por_sede[sede].append(nombre)

    # Resumen final
    print("\nResumen:")
    print("Inscripciones v√°lidas:", sum(len(v) for v in registro_por_sede.values()))
    print("Estudiantes por sede:")
    for sede, estudiantes in registro_por_sede.items():
        print(f"  - {sede}: {len(estudiantes)}")
    print("Sedes √∫nicas (ordenadas):", sorted(registro_por_sede.keys()))


# Ejemplo de entrada
inscripciones = [
    ("Ana", 21, "Centro"),
    ("Luis", 17, "Norte"),
    ("Ana", 21, "Centro"),
    ("Carlos", 15, "Sur"),
    ("Marta", 30, "Norte"),
    ("Pedro", "veinte", "Centro"),
    ("Elena", 25, 123)
]

# Ejecutar procesamiento
procesar_inscripciones(inscripciones)

````

```python
grupo_A = {'Ana', 'Luis', 'Carlos', 'Mar√≠a', 'Sof√≠a', 'Jorge', 'Isabel'}
grupo_B = {'Pedro', 'Mar√≠a', 'Luis', 'Diana', 'Isabel', 'Fernando', 'Jorge'}
```

Realiza las siguientes operaciones:

2. Verifica si `Ana` pertenece a `grupo_A` y si `Diana` no pertenece a `grupo_A`.

3. Crea un conjunto vac√≠o llamado `grupo_C` usando `set()`.

4. Agrega `Valeria` a `grupo_C` con `add()`.

5. Elimina `Valeria` usando `remove()`.

6. Intenta eliminar `Valeria` nuevamente con `discard()` (no debe imprimir error).

7. Agrega a `grupo_C` los nombres `'Mateo'`, `'Luis'`, `'Pedro'`, y luego elimina un elemento al azar usando `pop()`.

8. Vac√≠a el conjunto `grupo_C` usando `clear()`.

9. Verifica si `grupo_A` y `grupo_B` son iguales o diferentes.

14. Muestra la *diferencia sim√©trica* de `grupo_A` y `grupo_B`.



````python
# 1. Definir los conjuntos iniciales
grupo_A = {'Ana', 'Luis', 'Carlos', 'Mar√≠a', 'Sof√≠a', 'Jorge', 'Isabel'}
grupo_B = {'Pedro', 'Mar√≠a', 'Luis', 'Diana', 'Isabel', 'Fernando', 'Jorge'}

print("Grupo A:", grupo_A)
print("Grupo B:", grupo_B)

# 2. Verificar pertenencia
print("¬øAna pertenece a grupo_A?:", 'Ana' in grupo_A)
print("¬øDiana no pertenece a grupo_A?:", 'Diana' not in grupo_A)

# 3. Crear conjunto vac√≠o
grupo_C = set()
print("Grupo C inicial:", grupo_C)

# 4. Agregar 'Valeria'
grupo_C.add('Valeria')
print("Grupo C con Valeria:", grupo_C)

# 5. Eliminar 'Valeria'
grupo_C.remove('Valeria')
print("Grupo C tras eliminar Valeria:", grupo_C)

# 6. Intentar eliminar 'Valeria' con discard (no error)
grupo_C.discard('Valeria')
print("Grupo C tras discard de Valeria:", grupo_C)

# 7. Agregar varios nombres y eliminar uno al azar con pop()
grupo_C.update(['Mateo', 'Luis', 'Pedro'])
print("Grupo C con nuevos nombres:", grupo_C)
eliminado = grupo_C.pop()
print("Elemento eliminado con pop():", eliminado)
print("Grupo C tras pop:", grupo_C)

# 8. Vaciar el conjunto
grupo_C.clear()
print("Grupo C vac√≠o:", grupo_C)

# 9. Verificar igualdad entre grupo_A y grupo_B
print("¬øgrupo_A == grupo_B?:", grupo_A == grupo_B)

# 10. Verificar si grupo_A est√° contenido en grupo_B
print("¬øgrupo_A es subconjunto de grupo_B?:", grupo_A.issubset(grupo_B))

# 11. Uni√≥n
print("Uni√≥n de A y B:", grupo_A.union(grupo_B))

# 12. Intersecci√≥n
print("Intersecci√≥n de A y B:", grupo_A.intersection(grupo_B))

# 13. Diferencia
print("Diferencia A - B:", grupo_A.difference(grupo_B))

# 14. Diferencia sim√©trica
print("Diferencia sim√©trica A y B:", grupo_A.symmetric_difference(grupo_B))
````

### **Pr√°ctica 6. Crear una funci√≥n**

Debes crear una funci√≥n llamada `calcular_descuento_total` que simule una operaci√≥n de compra con descuento e impuesto. La funci√≥n y su uso deben cumplir los siguientes pasos:


1. Define la funci√≥n `calcular_descuento_total` que reciba **cuatro par√°metros**:

  - `precio_unitario` (float): precio de un solo producto.  
  - `cantidad` (int): n√∫mero de productos comprados.  
  - `impuesto` (float): porcentaje de impuesto a aplicar (por ejemplo, `15` representa un 15%).  
  - `descuento` (float, opcional): porcentaje de descuento a aplicar (por defecto debe ser `0`).



2. Dentro de la funci√≥n, realiza los siguientes c√°lculos:

  a. Calcula el **total bruto**: `precio_unitario * cantidad`  
  b. Calcula el **total con descuento**: aplica el descuento al total bruto  
  c. Calcula el **total final**: aplica el impuesto al total con descuento  


3. La funci√≥n debe:

  - **Retornar una tupla** con dos elementos:  
  - El total con descuento  
  - El total final con impuesto

4. Incluye un **docstring completo** dentro de la funci√≥n que explique:

  - Qu√© hace la funci√≥n  
  - Qu√© par√°metros recibe  
  - Qu√© tipo de datos retorna  


5. En el programa principal (fuera de la funci√≥n):

  a. Llama a la funci√≥n con valores concretos.  
  b. Guarda el resultado en dos variables: `subtotal` y `total`.  
  c. Imprime cada valor por separado.  

6. Luego, usa el `total` retornado para:

  - Compara el `total` con un presupuesto disponible (por ejemplo: `presupuesto = 160`)  
  - Imprimir un mensaje que indique si la compra est√° dentro del presupuesto o lo excede.

````python
def calcular_descuento_total(precio_unitario, cantidad, impuesto, descuento=0):
    """
    Calcula el subtotal con descuento y el total final con impuesto de una compra.

    Par√°metros:
    - precio_unitario (float): precio de un solo producto
    - cantidad (int): n√∫mero de productos comprados
    - impuesto (float): porcentaje de impuesto a aplicar (ejemplo: 15 representa 15%)
    - descuento (float, opcional): porcentaje de descuento a aplicar (por defecto 0)

    Retorna:
    - tuple: (total_con_descuento, total_final)
    """

    # Total bruto
    total_bruto = precio_unitario * cantidad

    # Aplicar descuento
    total_con_descuento = total_bruto * (1 - descuento / 100)

    # Aplicar impuesto
    total_final = total_con_descuento * (1 + impuesto / 100)

    return (total_con_descuento, total_final)

# Valores de ejemplo
precio_unitario = 50.0
cantidad = 3
impuesto = 15
descuento = 10

# Llamada a la funci√≥n
subtotal, total = calcular_descuento_total(precio_unitario, cantidad, impuesto, descuento)

# Mostrar resultados
print("Subtotal con descuento:", subtotal)
print("Total final con impuesto:", total)

# Comparar con presupuesto
presupuesto = 160
if total <= presupuesto:
    print("La compra est√° dentro del presupuesto.")
else:
    print("La compra excede el presupuesto.")
````

### **Pr√°ctica 9. Crear una clase**

1. Crea una clase llamada `Producto` que represente un producto con los siguientes elementos:
   - Un atributo `nombre`, que guarda el nombre del producto.
   - Un atributo `precio`, que guarda el precio sin IVA.

2. Define en la clase un m√©todo llamado `precio_con_iva()` que:
   - Calcule el precio final del producto aplicando un 19% de IVA.
   - Retorne el precio con IVA.

3. Crea una instancia de la clase `Producto` con los siguientes datos:
   - `nombre`: `"Camiseta"`
   - `precio`: `20000`

4. Imprime los atributos del objeto:
   - El nombre del producto.
   - El precio sin IVA.

5. Llama al m√©todo `precio_con_iva()` e imprime el resultado (el precio con IVA).

````python
# Definici√≥n de la clase Producto
class Producto:
    """
    Representa un producto con nombre y precio sin IVA.

    M√©todos:
    - precio_con_iva(): calcula el precio final aplicando un 19% de IVA.
    """

    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

    def precio_con_iva(self):
        return self.precio * 1.19


# Crear una instancia de la clase Producto
producto1 = Producto("Camiseta", 20000)

# Imprimir atributos del objeto
print("Nombre del producto:", producto1.nombre)
print("Precio sin IVA:", producto1.precio)

# Llamar al m√©todo precio_con_iva() e imprimir el resultado
print("Precio con IVA:", producto1.precio_con_iva())
````
