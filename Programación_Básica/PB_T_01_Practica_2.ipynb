<h2 style="color:#2E4053; font-family:Georgia; border-left:6px solid #2E86C1; padding-left:10px;">
üìä Caso 1: Problema: Validaci√≥n datos de ventas
</h2>


````python
# Lista de registros de ejemplo
registros = [
    (200, 50),
    ("abc", 10),
    (-10, 20),
    (300, -5),
    (400, 80)
]

# Lista para guardar registros v√°lidos
registros_validos = []

# Procesar cada registro
for registro in registros:
    usuarios, items = registro

    # Validar que ambos sean enteros
    if not (isinstance(usuarios, int) and isinstance(items, int)):
        print("Registro no num√©rico:", registro)
        continue

    # Validar rango de valores
    if usuarios <= 0 or items < 0:
        print("Registro fuera de rango:", registro)
        continue

    # Si pasa todas las validaciones, guardar
    registros_validos.append(registro)

# Resumen final
print("\nResumen:")
print("Registros v√°lidos:", len(registros_validos))

if registros_validos:  # Solo si hay registros v√°lidos
    # M√°ximo y m√≠nimo de usuarios
    max_usuarios = max(r[0] for r in registros_validos)
    min_usuarios = min(r[0] for r in registros_validos)

    # Promedio de √≠tems vendidos
    promedio_items = sum(r[1] for r in registros_validos) / len(registros_validos)

    print("M√°x usuarios:", max_usuarios)
    print("M√≠n usuarios:", min_usuarios)
    print("Promedio √≠tems vendidos:", promedio_items)
````

### **Pr√°ctica 1. Sentencia while**

Escriba un c√≥digo en Python que use un bucle `while` para hacer lo siguiente:

* Inicialice una variable llamada `x` con el valor `1`.

* Inicialice otra variable llamada `suma` con el valor `0`.

* Use un bucle while para sumar los n√∫meros del 1 al 5.

* En cada paso, vaya actualizando el valor de suma.

* Al finalizar el ciclo, use `print()` para mostrar el resultado final con un mensaje como: `La suma total es: 15` y `Fin del ciclo`


<br>

````python
# Inicializar variables
x = 1
suma = 0

# Bucle while para sumar del 1 al 5
while x <= 5:
    suma += x   # Actualizar la suma
    x += 1      # Incrementar x en cada paso

# Mostrar resultado final
print("La suma total es:", suma)
print("Fin del ciclo")
````

### **Pr√°ctica 2. Sentencia while**

A continuaci√≥n se presenta un programa que simula la lectura de datos desde 3 sensores.

Cada sensor realiza 4 intentos de lectura.

Completa el c√≥digo para que:

* Se repitan los intentos de lectura por cada sensor.

* Se calcule un valor simulado (dato = sensor * intento - 5).

* Si el valor es negativo, se considere un error y se cuente.

* Imprimir en pantalla los errores detectados durante la simulaci√≥n.

````python
# Inicializar contador de errores
errores = 0

# Simular lectura de 3 sensores
sensor = 1
while sensor <= 3:
    intento = 1
    while intento <= 4:
        # Calcular valor simulado
        dato = sensor * intento - 5

        # Validar si es error
        if dato < 0:
            print(f"Error detectado en sensor {sensor}, intento {intento}: dato={dato}")
            errores += 1

        intento += 1
    sensor += 1

# Mostrar resumen final
print("\nTotal de errores detectados:", errores)
````

## **Pr√°ctica 3. Sentencia continue**

Una estaci√≥n meteorol√≥gica est√° revisando una lista de temperaturas registradas cada hora durante el d√≠a (simuladas del 24 al 0).

Las temperaturas negativas se consideran errores del sensor, por lo que deben ignorarse usando la instrucci√≥n continue.

Tu programa debe:

* Iniciar con `hora = 24` y descender hasta `0`.

* Generar una temperatura simulada con la f√≥rmula:
`temperatura = (hora - 12) * 1.5`

* Si la temperatura es negativa, usar `continue` para saltar ese caso.

* Si no debe imprimir la hora y la temperatura.

* Al final, mostrar `"Fin del monitoreo."`




````python
hora = 24
while hora >= 0:
  temperatura = (hora - 12) * 1.5

  if temperatura < 0:
    hora -=1
    continue

  else:
    print(f"Hora: {hora}\tTemperatura: {temperatura:.2f}")
    hora -=1

print("fin del monitoreo")
````

Utiliza un ciclo `while True` para simular este crecimiento.
Cuando el tama√±o supere el l√≠mite, imprime un mensaje de alerta y termina el ciclo usando `break`.

````python
tama√±o = 50
while True:
  if tama√±o > 1000:
    print(f"Tama√±o actual: {tama√±o} MB")
    print("alerta archivo supera el limite de almacenamiento")
    break
  else:
    print(f"Tama√±o actual: {tama√±o} MB")
    tama√±o = tama√±o * 2
````

2. Eliminar elementos

3. Filtrar y transformar
* Dada la lista [3, -5, 7, -2, 0, 12, -9], crea:

* Una nueva lista con los elementos positivos.

* Una lista con el cuadrado de los negativos usando comprensi√≥n de listas.

4. Ordenar y analizar
* Ordena la lista [12, 5, 7, 3, 19, 1] de mayor a menor.

* Encuentra el valor m√°ximo y m√≠nimo.

* Cuenta cu√°ntas veces aparece el n√∫mero 3 en la lista [3, 3, 1, 5, 3, 2].

* Crea una copia de esa lista y luego invierte el orden de la copia.

5. Genera una lista por comprensi√≥n con los cuadrados de los n√∫meros del 1 al 20.

````python
# Crear y modificar una lista
numeros = list(range(1, 11))   # [1,2,...,10]
print("Lista inicial:", numeros)

# Cambiar el valor en la posici√≥n 0 por 100
numeros[0] = 100

# Multiplicar la lista por 2 (repetici√≥n)
numeros = numeros * 2

# Agregar el valor 200 al final con append
numeros.append(200)

# Insertar el valor 50 en la posici√≥n 3
numeros.insert(3, 50)

print("Lista modificada:", numeros)

# Eliminar elementos
numeros.remove(100)  # elimina la primera aparici√≥n de 100
eliminado = numeros.pop()  # elimina el √∫ltimo elemento y lo guarda

print("Lista despu√©s de eliminar:", numeros)
print("Valor eliminado:", eliminado)


# Filtrar y transformar
lista = [3, -5, 7, -2, 0, 12, -9]

positivos = [x for x in lista if x > 0]
cuadrados_negativos = [x**2 for x in lista if x < 0]

print("Positivos:", positivos)
print("Cuadrados de negativos:", cuadrados_negativos)


# Ordenar y analizar
lista2 = [12, 5, 7, 3, 19, 1]

# Ordenar de mayor a menor
lista2.sort(reverse=True)
print("Lista ordenada de mayor a menor:", lista2)

# M√°ximo y m√≠nimo
print("M√°ximo:", max(lista2))
print("M√≠nimo:", min(lista2))


# Contar apariciones
lista3 = [3, 3, 1, 5, 3, 2]
print("N√∫mero de veces que aparece 3:", lista3.count(3))

# Copia e invertir
copia = lista3.copy()
copia.reverse()
print("Copia invertida:", copia)


# Lista por comprensi√≥n con cuadrados del 1 al 20
cuadrados = [x**2 for x in range(1, 21)]
print("Cuadrados del 1 al 20:", cuadrados)
````
